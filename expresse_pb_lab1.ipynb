Sets and ParametersSets:$I$: 

Set of bricks, $I = \{1, 2, ..., 22\}$.$J$: 

Set of Sales Representatives (SRs), $J = \{1, 2, 3, 4\}$.

Parameters:

$w_i$: The workload (index value) of brick $i \in I$ (from Table 2 2).

$d_{ij}$: The travel distance between the office of SR $j \in J$ and brick $i \in I$ (from Table 3).

$A_{ij}^{current}$: The current assignment. Let's define this as $1$ if brick $i$ is currently assigned to SR $j$, and $0$ otherwise (from Table 1 4).

$L_j, U_j$: The lower and upper workload bounds for SR $j \in J$. For the first run, $L_j = 0.8$ and $U_j = 1.2$ for all $j$5.

Decision Variable$x_{ij}$: 

A binary variable, where:

$x_{ij} = 1$ if brick $i$ is assigned to SR

 $j$.$x_{ij} = 0$ otherwise.

1. Brick AssignmentEach brick $i$ must be assigned to exactly one SR $j$1.$$\sum_{j \in J} x_{ij} = 1 \quad \forall i \in I$$

The total workload for each SR $j$ (the sum of index values of bricks assigned to them) must remain within the specified bounds 2.Lower Bound:$$\sum_{i \in I} w_i \cdot x_{ij} \ge L_j \quad \forall j \in J$$Upper Bound:$$\sum_{i \in I} w_i \cdot x_{ij} \le U_j \quad \forall j \in J$$For the first part of the project, these bounds are set as $L_j = 0.8$ and $U_j = 1.2$ for all SRs3.
